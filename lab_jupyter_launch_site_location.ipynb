```python
# TASK 3: Calculate the distances between a launch site to its proximities
import folium
from folium.plugins import MarkerCluster
from math import sin, cos, sqrt, atan2, radians
def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the distance between two points on Earth using Haversine formula
    Returns distance in kilometers
    """
    R = 6371  # Earth radius in km

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c
launch_site_name = "CCAFS LC-40"
launch_site_lat = 28.562302
launch_site_lon = -80.577356
# Coastline
coast_lat, coast_lon = 28.563, -80.567

# Railway
rail_lat, rail_lon = 28.573, -80.585

# Highway
road_lat, road_lon = 28.570, -80.650

# City (Orlando)
city_lat, city_lon = 28.538336, -81.379234
distance_to_coast = calculate_distance(
    launch_site_lat, launch_site_lon, coast_lat, coast_lon
)

distance_to_rail = calculate_distance(
    launch_site_lat, launch_site_lon, rail_lat, rail_lon
)

distance_to_road = calculate_distance(
    launch_site_lat, launch_site_lon, road_lat, road_lon
)

distance_to_city = calculate_distance(
    launch_site_lat, launch_site_lon, city_lat, city_lon
)
print(f"Launch Site: {launch_site_name}")
print(f"Distance to coastline: {distance_to_coast:.2f} km")
print(f"Distance to railway: {distance_to_rail:.2f} km")
print(f"Distance to highway: {distance_to_road:.2f} km")
print(f"Distance to nearest city: {distance_to_city:.2f} km")
# Create map centered on launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Add launch site marker
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup="Launch Site: CCAFS LC-40",
    icon=folium.Icon(color='red', icon='rocket', prefix='fa')
).add_to(site_map)

# Add proximity markers
folium.Marker(
    location=[coast_lat, coast_lon],
    popup=f"Coastline ({distance_to_coast:.2f} km)",
    icon=folium.Icon(color='blue')
).add_to(site_map)

folium.Marker(
    location=[rail_lat, rail_lon],
    popup=f"Railway ({distance_to_rail:.2f} km)",
    icon=folium.Icon(color='green')
).add_to(site_map)

folium.Marker(
    location=[road_lat, road_lon],
    popup=f"Highway ({distance_to_road:.2f} km)",
    icon=folium.Icon(color='purple')
).add_to(site_map)

folium.Marker(
    location=[city_lat, city_lon],
    popup=f"City ({distance_to_city:.2f} km)",
    icon=folium.Icon(color='orange')
).add_to(site_map)

site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# STEP 1: Import libraries and define distance function
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c


# STEP 2: Define launch site coordinates
launch_site_lat = 28.562302
launch_site_lon = -80.577356


# STEP 3: Define closest city, railway, and highway coordinates
city_lat, city_lon = 28.538336, -81.379234
railway_lat, railway_lon = 28.572872, -80.585269
highway_lat, highway_lon = 28.565500, -80.590000


# STEP 4: Calculate distances
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)


# STEP 5: Create base map
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)


# STEP 6: Create distance markers
city_marker = folium.Marker(
    location=[city_lat, city_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size:12px;color:#2980b9;"><b>{distance_city:.2f} KM (City)</b></div>'
    )
)

railway_marker = folium.Marker(
    location=[railway_lat, railway_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size:12px;color:#27ae60;"><b>{distance_railway:.2f} KM (Railway)</b></div>'
    )
)

highway_marker = folium.Marker(
    location=[highway_lat, highway_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size:12px;color:#c0392b;"><b>{distance_highway:.2f} KM (Highway)</b></div>'
    )
)

site_map.add_child(city_marker)
site_map.add_child(railway_marker)
site_map.add_child(highway_marker)


# STEP 7: Draw lines between launch site and proximities
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [city_lat, city_lon]],
    weight=2, color='blue'
).add_to(site_map)

folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]],
    weight=2, color='green'
).add_to(site_map)

folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]],
    weight=2, color='red'
).add_to(site_map)


# Display the map
site_map
```
